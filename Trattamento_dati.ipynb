```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit as cf
```

```python
def linear_model(x, m, b):
    return m * x + b
```

```python
data = np.loadtxt(r"C:\Users\Mattia\Lab_Fisica_2\Book1.txt", delimiter='\t', skiprows=1)


x = data[:, 0]
y = data[:, 1]
x_error = data[:, 2]
y_error = data[:, 3]
```

```python
params, params_covariance = cf(linear_model, x, y, sigma=y_error, absolute_sigma=True)
m_fit, b_fit = params
```

```python
plt.errorbar(x, y, xerr=x_error, yerr=y_error, fmt='o', label="Data with error bars", capsize=2)
plt.plot(x, linear_model(x, *params), color='red', label=f"Best fit: y = {m_fit:.2f}x + {b_fit:.2f}")
plt.title("Best Fit with Error Bars")
plt.xlabel("Massa Axis")
plt.ylabel("Volume Axis")
plt.legend()
plt.grid(True)
```
