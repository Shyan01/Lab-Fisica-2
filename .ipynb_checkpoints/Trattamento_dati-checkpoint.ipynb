```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('Book1.txt',delimiter = None,  skiprows = 1)

x = data[:, 0]
y = data[:, 1]
x_error = data[:, 2]
y_error = data[:, 3]
```

```python
plt.errorbar(x,y, xerr = x_error,yerr = y_error, fmt='.', capsize = 5, label = "Punti con errore")
plt.title("Grafico con errori")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```
