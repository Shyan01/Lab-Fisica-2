```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def Rend_Ideale (Tf, Tc):  #Definisco la funzione per il cop Ideale

    cop = Tf/(Tc - Tf)
    return cop
```

```python
def Rend_Reale (lamnda, delta, S, h, I, V):

    cop = (lamnda* (delta)*S)/(I*V*h)
    return cop
```

```python
def Errore_Generatore(V):

    error_v = 0.01 + 0.03*V
    return error_v
```
